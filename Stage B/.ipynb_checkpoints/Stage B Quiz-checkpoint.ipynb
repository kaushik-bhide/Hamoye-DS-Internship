## Q12) 

```python
import pandas as pd
energy = pd.read_csv('energy.csv',parse_dates=['date'])
energy = energy.drop(['date','lights'],axis=1)
energy.head()
```

```python
energy.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression,Ridge,Lasso
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(energy), columns=energy.columns)
features_df = normalised_df[['T2']]
heating_target = normalised_df[['T6']]
```

```python
#Now, we split our dataset into the training and testing dataset. Recall that we had earlier segmented the features and target variables.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, heating_target, test_size=0.30, random_state=42)
```

```python
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train, y_train)
#obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
print("The R2 score is:",r2_score)
print("The Rounded R2 Score is:",round(r2_score))
```

## Q13)

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression,Ridge,Lasso
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(energy), columns=energy.columns)
features_df = normalised_df
heating_target = normalised_df['Appliances']
```

```python
#Now, we split our dataset into the training and testing dataset. Recall that we had earlier segmented the features and target variables.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df, heating_target, test_size=0.30, random_state=42)
```

```python
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train, y_train)
#obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
import numpy as np
predicted_values = np.around(predicted_values,6)
mae = mean_absolute_error(y_test, predicted_values)
print("{:.2f}".format(float(mae)))
```

## Q14)

```python
import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
print("{:.2f}".format(float(rss)))
```

## Q15)

```python
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
print("{:.3f}".format(float(rmse)))
```

## Q16)

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
print("The R2 score is:",r2_score)
```

## Q17)

```python
#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
  #this function returns the weight of every feature
    model.fit(x_train,y_train)
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
model = LinearRegression()
lasso_reg = Lasso(alpha=0.001)
ridge_reg = Ridge(alpha=0.001)
linear_model_weights = get_weights_df(model, x_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')


final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```
